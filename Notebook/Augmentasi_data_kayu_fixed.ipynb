```python
import os
import cv2
from google.colab import drive
import imgaug as ia
from imgaug import augmenters as iaa
import numpy as np
```

```python
# Inisialisasi objek augmentasi
aug = iaa.Fliplr(1.0)

aug1 = iaa.Affine(
        rotate=(90),
        mode="edge",
        cval=255)

aug2 = iaa.Flipud(1.0)
```

```python
# Mendefinisikan path ke direktori gambar asli dan direktori tujuan augmentasi
input_path = "wood_v6/test_data/nantu"
output_path = "wood_augmented2/test_data/nantu"

# Membuat direktori tujuan jika belum ada
if not os.path.exists(output_path):
    os.makedirs(output_path)

# Inisialisasi variabel penghitung jumlah data
counter = 0

# Loop melalui setiap gambar di direktori input
for i, image_file_name in enumerate(sorted(os.listdir(input_path))):
    # Membaca gambar
    image = cv2.imread(os.path.join(input_path, image_file_name))

    # Melakukan augmentasi
    images_aug = aug(images=[image])
    images_aug1 = aug1(images=[image])
    images_aug2 = aug2(images=[image])


    # Menyimpan gambar hasil augmentasi
    for j, image_aug in enumerate(images_aug):
        new_image_file_name = f"nantu_test_{i * 3 + 1 :03d}.jpg"
        image_aug_np = np.asarray(image_aug)
        output_file_path = os.path.join(output_path, new_image_file_name)
        success = cv2.imwrite(output_file_path, image_aug_np)

        # Memeriksa status hasil penyimpanan
        if success:
            counter += 1  # Menginkrementasi counter
            print(f"Augmentasi berhasil: {output_file_path}")
        else:
            print(f"Augmentasi gagal: {output_file_path}")

    # Menyimpan gambar hasil augmentasi
    for j, image_aug in enumerate(images_aug1):
        new_image_file_name = f"nantu_test_{i * 3 + 2 :03d}.jpg"
        image_aug_np = np.asarray(image_aug)
        output_file_path = os.path.join(output_path, new_image_file_name)
        success = cv2.imwrite(output_file_path, image_aug_np)

        # Memeriksa status hasil penyimpanan
        if success:
            counter += 1  # Menginkrementasi counter
            print(f"Augmentasi berhasil: {output_file_path}")
        else:
            print(f"Augmentasi gagal: {output_file_path}")

    # Menyimpan gambar hasil augmentasi
    for j, image_aug in enumerate(images_aug2):
        new_image_file_name = f"nantu_test_{i * 3 + 3 :03d}.jpg"
        image_aug_np = np.asarray(image_aug)
        output_file_path = os.path.join(output_path, new_image_file_name)
        success = cv2.imwrite(output_file_path, image_aug_np)

        # Memeriksa status hasil penyimpanan
        if success:
            counter += 1  # Menginkrementasi counter
            print(f"Augmentasi berhasil: {output_file_path}")
        else:
            print(f"Augmentasi gagal: {output_file_path}")

# Menampilkan jumlah keseluruhan data hasil augmentasi
print(f"Jumlah data hasil augmentasi: {counter}")
```
